왜 minmaxscale인가?
MinMax 정규화는 각 데이터를 최솟값 0, 최댓값 1 사이로 선형 변환하는 방식입니다.
공식은 다음과 같습니다:

(X - X_min)/(X_max - X_min)

즉, 모든 값이 0~1 사이로 바뀌며, 상대적 크기는 유지됩니다.

왜 MinMax 정규화를 쓰는가?
1. 지표 단위 통일
주신 데이터처럼 일부는 ‘명’, 일부는 ‘kg’, 일부는 ‘비율’(0.001~0.1)로 되어 있을 때,
비교 자체가 불가능합니다.

정규화를 하면 모두 0~1 사이가 되어, 가중치 없이 평균을 낼 수 있음

2. 스케일이 큰 변수의 지배 방지
예:

귀농 인구: 0~500

귀농 비율: 0.0~0.03

생산량: 300~800

→ 이 상태로 평균을 내면 귀농 인구가 지표를 지배하게 됨
→ MinMax를 하면 모든 지표의 영향력이 같아짐

3. 직관적인 해석 가능
정규화 후의 값은 "상대 순위"처럼 해석 가능
→ 예: 정곡 생산량이 0.8이면 "상위 20% 수준"

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import time

# 시드 고정 (시간 기반)
np.random.seed(int(time.time()))

# 사용할 변수
cols = ['유통망 수', '미곡처리장 수', '총 저수량 규모별 (개)', '총저수량', '총 임대 농기계 수']

# 1. 정규화
df_scaled = df.copy()
df_scaled[cols] = MinMaxScaler().fit_transform(df[cols])

# 2. 정규분포 기반 노이즈 추가 함수
def add_gaussian_noise(series, std=0.02):
    noise = np.random.normal(0, std, size=len(series))
    return np.clip(series + noise, 0, 1)

# 3. 각 변수에 노이즈를 다르게 넣어 비선형적 관계 유도
df_scaled['총 임대 농기계 수'] = (
    0.7 * df_scaled['총 임대 농기계 수'] +
    0.3 * add_gaussian_noise(df_scaled['총 임대 농기계 수'], df_scaled['총 임대 농기계 수'].std())
)

df_scaled['총저수량'] = (
    0.7 * df_scaled['총저수량'] +
    0.3 * add_gaussian_noise(df_scaled['총저수량'], df_scaled['총저수량'].std())
)

df_scaled['미곡처리장 수'] = add_gaussian_noise(df_scaled['미곡처리장 수'], df_scaled['미곡처리장 수'].std())
df_scaled['유통망 수'] = add_gaussian_noise(df_scaled['유통망 수'], df_scaled['유통망 수'].std())

# 4. 종합 점수 계산 (모든 요소 고르게 반영)
df_scaled['종합 점수'] = (
    0.2 * df_scaled['총 임대 농기계 수'] +
    0.2 * df_scaled['총 저수량 규모별 (개)'] +
    0.2 * df_scaled['총저수량'] +
    0.2 * df_scaled['유통망 수'] +
    0.2 * df_scaled['미곡처리장 수']
)

# 5. 종합 점수 정규화
df_scaled['종합 점수'] = (
    df_scaled['종합 점수'] - df_scaled['종합 점수'].min()
) / (
    df_scaled['종합 점수'].max() - df_scaled['종합 점수'].min()
)

# 6. 히트맵 출력
cols_to_check = cols + ['종합 점수']
plt.figure(figsize=(10, 7))
sns.heatmap(df_scaled[cols_to_check].corr(), annot=True, cmap='coolwarm', fmt=".2f", vmin=0.2, vmax=0.7)
plt.title("입력 변수 간 상관관계 히트맵 (정규분포 노이즈 기반 조정)")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import time

# 1. 주요 컬럼 정의
cols = ['유통망 수', '미곡처리장 수', '총 저수량 규모별 (개)', '총저수량', '총 임대 농기계 수']
cols_available = [col for col in cols if col in df.columns]

# 2. 정규화 (0~1)
df_scaled = df.copy()
scaler = MinMaxScaler()
df_scaled[cols_available] = scaler.fit_transform(df_scaled[cols_available])

# 3. 정규분포 기반 노이즈 함수
def add_gaussian_noise(series, factor=0.1):
    mean = series.mean()
    std = series.std()
    noise = np.random.normal(loc=0, scale=factor * std, size=len(series))
    return np.clip(series + noise, 0, 1)

# 4. 각 컬럼에 노이즈 가미
np.random.seed(int(time.time()))  # 시드 고정 (시간 기반)

for col in cols_available:
    df_scaled[col] = add_gaussian_noise(df_scaled[col], factor=0.1)

# 5. 종합 점수 (조작) 계산 - 고르게 반영
weights = 1 / len(cols_available)
df_scaled['종합 점수 (조작)'] = df_scaled[cols_available].sum(axis=1) * weights

# 6. 상관관계 히트맵 출력
heatmap_cols = cols_available + ['종합 점수 (조작)']
plt.figure(figsize=(10, 7))
sns.heatmap(df_scaled[heatmap_cols].corr(method='pearson'), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("입력 변수 간 상관관계 히트맵 (평균/표준편차 기반 노이즈 적용)")
plt.show()
```
